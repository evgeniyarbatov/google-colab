```python
import os
import pandas as pd

def convert_csv_to_parquet(directory):
    # Ensure directory path ends without a trailing slash
    directory = directory.rstrip("/\\")

    for filename in os.listdir(directory):
        if filename.lower().endswith(".csv"):
            csv_path = os.path.join(directory, filename)
            parquet_path = os.path.join(directory, filename[:-4] + ".parquet")

            print(f"Converting: {csv_path} -> {parquet_path}")

            df = pd.read_csv(csv_path)
            df.to_parquet(parquet_path, index=False)

    print("Conversion completed.")

# Example usage:
convert_csv_to_parquet("/content/")
```
