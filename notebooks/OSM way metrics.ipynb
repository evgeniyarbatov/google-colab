```python
import pandas as pd

# Load the dataframe.
df = pd.read_csv('/content/metrics.csv')

# Count unique way_id and unique name.
unique_way_ids = df['way_id'].nunique()
unique_names = df['name'].nunique()

# Print the results.
print(f"Number of unique way_ids: {unique_way_ids}")
print(f"Number of unique names: {unique_names}")
```

```python
# Filter the DataFrame to include only rows where the 'name' is 'unknown'
unknown_names_df = df[df['name'] == 'unnamed']

# Count the number of unique way_ids in the filtered DataFrame
unique_unknown_way_ids = unknown_names_df['way_id'].nunique()

# Print the result
print(f"Number of unique way_ids with 'unnamed' name: {unique_unknown_way_ids}")
```
