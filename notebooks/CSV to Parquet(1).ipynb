```python
import pandas as pd
import numpy as np
import sys

csv_path = '/content/MetObjects.csv'
parquet_path = '/content/MetObjects.parquet'

def csv_to_parquet(csv_path: str, parquet_path: str):
    print(f"Reading CSV: {csv_path}")
    df = pd.read_csv(csv_path, on_bad_lines="skip")

    # Convert all object columns to string to avoid Arrow conversion errors
    for col in df.columns:
        if df[col].dtype == "object":
            print(f"Converting to string: {col}")
            df[col] = df[col].astype(str)

    print(f"Writing Parquet: {parquet_path}")
    df.to_parquet(parquet_path, index=False)
    print("Done!")

csv_to_parquet(csv_path, parquet_path)
```
