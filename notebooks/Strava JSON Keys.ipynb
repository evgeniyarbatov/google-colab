```python
import json
from collections import defaultdict, Counter

file_path = '/content/strava_activities.json'

# Load JSON
with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

def extract_values(obj, parent_key='', values_dict=None):
    """Recursively extract all values per key from JSON."""
    if values_dict is None:
        values_dict = defaultdict(list)

    if isinstance(obj, dict):
        for k, v in obj.items():
            full_key = f"{parent_key}.{k}" if parent_key else k
            values_dict[full_key].append(v)
            extract_values(v, full_key, values_dict)
    elif isinstance(obj, list):
        for item in obj:
            extract_values(item, parent_key, values_dict)
    return values_dict

# Extract values per key
values_per_key = extract_values(data)

# Count unique values for each key
unique_counts = {key: len(set(map(lambda x: json.dumps(x, sort_keys=True), vals)))
                 for key, vals in values_per_key.items()}

# Sort keys by number of unique values (most frequent on top)
sorted_keys = sorted(unique_counts.items(), key=lambda x: x[1], reverse=True)

# Print
for key, count in sorted_keys:
    print(f"{key}: {count}")
```
