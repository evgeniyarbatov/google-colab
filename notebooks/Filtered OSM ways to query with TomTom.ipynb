```python
import pandas as pd

ways_df = pd.read_csv('/content/ways.csv')

ways_below_30k = ways_df[ways_df['distance'] < 30000]

num_ways_below_30k = len(ways_below_30k)

print(f"The number of ways with a distance below 30000 is: {num_ways_below_30k}")
```

```python
ways_df_cleaned = ways_df.dropna(subset=['min_distance'])
ways_df_cleaned['min_distance_km'] = (ways_df_cleaned['min_distance'] / 1000).round(1)
way_id_by_min_distance_km = ways_df_cleaned.groupby('min_distance_km')['way_id'].count()
with pd.option_context('display.max_rows', None):
    display(way_id_by_min_distance_km)
```

```python
ways_filtered = ways_df_cleaned[(ways_df_cleaned['min_distance_km'] > 0.6) & (ways_df_cleaned['distance'] < 30000)]
num_ways_filtered = len(ways_filtered)

print(f"The number of ways with min_distance_km greater than 0.1 and distance below 30000 is: {num_ways_filtered}")
```

```python
import folium

# Get the center of the filtered data
center_lat = ways_filtered['lat'].mean()
center_lon = ways_filtered['lon'].mean()

# Create a map centered around the data
m = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# Add markers for each way in the filtered data
for index, row in ways_filtered.iterrows():
    folium.Marker([row['lat'], row['lon']]).add_to(m)

# Display the map
m
```
